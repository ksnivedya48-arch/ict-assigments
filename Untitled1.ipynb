```python
quiz = {
    "easy": [("2+2?", "4")],
    "medium": [("len('Python')?", "6")],
    "hard": [("bool('False')?", "True")]
}

score_rule = {
    "easy": (1, -0.25),
    "medium": (2, -0.5),
    "hard": (3, -1)
}

name = input("Name: ")
score = 0
seed = len(name)

for level in quiz:
    q, ans = quiz[level][seed % len(quiz[level])]

    print("\n", level.upper(), ":", q)
    user = input("Answer: ")

    if user == ans:
        score += score_rule[level][0]
        print("Correct")
    else:
        score += score_rule[level][1]
        print("Wrong")

print("\nFinal Score:", score)
```

Question number 3

```python
store = {
    "Pen": [10, 20],
    "Book": [50, 10]
}

cart = []
total = 0

while True:
    item = input("Item (or done): ")
    if item == "done":
        break

    if item in store:
        qty = int(input("Qty: "))

        if qty <= store[item][1]:
            cart.append((item, qty))
            store[item][1] -= qty
        else:
            print("Low stock!")
    else:
        print("Not available")

print("\n--- BILL ---")
for item, qty in cart:
    cost = store[item][0] * qty
    total += cost
    print(item, "x", qty, "=", cost)

tax = total * 0.05
discount = total * 0.1 if total > 200 else 0
pay = total + tax - discount

print("Subtotal:", total)
print("Tax:", tax)
print("Discount:", discount)
print("Payable:", pay)

print("\nLow Stock:")
for i in store:
    if store[i][1] < 5:
      print(i, "running low")
```
